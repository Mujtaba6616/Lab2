```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
image = Image.open('download.jpg')

# summarize some details about the image

array=np.array(image)
print(array)
plt.imshow(array)
plt.show()


temp=np.rot90(array)
plt.imshow(temp)
plt.show()


temp1=np.fliplr(temp)
plt.imshow(temp1)
plt.show()

gray_img = np.dot (array[..., :3], [0.299, 0.587, 0.114])
plt.imshow(gray_img, cmap='gray')
plt.show()



```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt  # Keep this import
import seaborn as sns
# Load the Iris dataset
iris = load_iris()
# Accessing the features (data) using NumPy array
X = np.array(iris.data) # (Features (sepal length, sepal width, petal length, petal
 #Accessing the target labels (species)
Y = np.array(iris.target) # Target variable (species: 0 for setosa, 1 for versicolor, 2

MEAN=np.mean(X,axis=0)
print('Mean of X:',MEAN)
MEDIAN=np.median(X,axis=0)
print(MEDIAN)

STD=np.std(X,axis=0)
print(STD)


MEAN1=np.mean(Y,axis=0)
print(MEAN1)
MEDIAN1=np.median(Y,axis=0)
print(MEDIAN1)

STD1=np.std(Y,axis=0)
print(STD1)


min1=np.min(X,axis=0)
max1=np.max(X,axis=0)
print('Minimum X: Maximum X',min1,max1)

min2=np.min(Y,axis=0)
max2=np.max(Y,axis=0)
print('Minimum Y and Maximum Y',min2,max2)


SepalLength = X[:,0]
SepalWidth = X[:, 1]
PetalLength = X[:, 2]
PetalWidth = X[:, 3]

plt.scatter(SepalLength,SepalWidth)
plt.show()

plt.hist(SepalLength)
plt.show()

sns.lineplot(x=np.arange(len(PetalLength)), y=PetalWidth)  # Assuming you want to plot SepalLength against its index
plt.show()
```
